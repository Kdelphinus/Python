```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv('data/movie_metadata.csv')
df.head()
```

## 문제1

```python
df.plot(kind = 'scatter', x = 'budget', y = 'imdb_score')
```

```python
q1 = df['budget'].quantile(0.25)
q3 = df['budget'].quantile(0.75)
iqr = q3 - q1
```

```python
condition = df['budget'] > 5 * iqr + q3
df.drop(df[condition].index, inplace=True)
```

```python
df.plot(kind = 'scatter', x = 'budget', y = 'imdb_score')
```
