```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data/titanic.csv')
df.head(3)
```

생존 여부는 'Survived'(1은 생존, 0은 사망)  
좌석 등급은 'Pclass'(1은 1등실, 2는 2등실, 3은 3등실)  
지불한 요금은 'Fare'

## 타이타닉의 승객은 30대와 40대가 가장 많다

```python
df['Age'].plot(kind = 'hist')
```

20대, 30대가 가장 많다

## 가장 높은 요금을 낸 사람은 30대이다

```python
df.sort_values(by = 'Fare', ascending=False).head()
```

가장 비싼 금액을 낸 사람은 30대이다

### 모범답안

```python
df.plot(kind = 'scatter', x = 'Age', y = 'Fare')
```

## 생존자가 사망자보다 더 많다

```python
df['Survived'].value_counts()
```

사망자가 더 많다

## 1등실, 2등실, 3등실 중 가장 많은 사람이 탑승한 곳은 3등실이다

```python
df['Pclass'].value_counts()
```

3등실 > 1등실 > 2등실

## 가장 생존율이 높은 객실 등급은 1등실이다

```python
sns.violinplot(data = df, x = 'Pclass', y = 'Survived')
```

가장 생존율이 높은 객실은 1등실이다

## 모범답안

```python
sns.kdeplot(x = df['Pclass'], y = df['Survived'])
```

3등급은 아래쪽이 밀집되어있다.  
그에 반해 1등급은 위쪽이 더 밀집되어 있다

## 나이가 어릴수록 생존율이 높다

## 모범답안

```python
sns.violinplot(data = df, x ='Survived', y = 'Age')
```

생존자(1)와 사망자(0)의 분포가 거의 유사하기에 큰 차이를 찾기 어렵다

## 나이보다 성별이 생존율에 더 많은 영향을 미친다

```python
sns.stripplot(data = df, x = 'Survived', y = 'Age', hue = 'Sex')
```

```python
sns.catplot(data = df, x = 'Survived', y = 'Age', hue = 'Sex', kind = 'swarm')
```

나의의 차이는 생존자와 사망자 경향이 비슷해 보이지만  
성별은 많이 차이나는 것을 볼 수 있다
