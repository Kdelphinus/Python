```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data/body.csv', index_col=0)
df.head()
```

```python
df['Height']
```

```python
df['Height'].value_counts().sort_index().plot()
```

```python
sns.kdeplot(df['Height'])
```

```python
sns.kdeplot(df['Height'], bw_adjust= 0.05)
```

```python
sns.kdeplot(df['Height'], bw_method= 0.05)
```

bw_method와 bw_adjust 모두 KDE를 얼마나 추측할 것인지 정하는 값이다
그러나 bw_adjust는 1이 가장 이상적인 반면 bw_method는 0.5만 되도 이미 이상적이다

기본적으론 bw_adjust를 이용하는 것이 좋다
