```python
import pandas as pd
```

```python
liverpool_df = pd.read_csv('data/liverpool.csv', index_col = 0)
liverpool_df
```

```python
# 기존의 데이터 프레임을 바꾸지 않고 새로운 데이터를 만들어버림
liverpool_df.rename(columns = {'position': 'Position'}) # {원래이름: 새로운 이름}
```

```python
# 한 번의 여러 개 바꾸기
liverpool_df.rename(columns = {
    'position': 'Position',
    'born': 'Born',
    'number': 'Number',
    'nationality': 'Nationality'
}, inplace = True)

liverpool_df
```

```python
liverpool_df.index.name = 'Player Name'
liverpool_df
```

```python
liverpool_df.set_index('Number')  # 기존 자료는 수정하지 않는다, 원래 인덱스는 날라가 버린다
```

```python
# index를 바꾸기 전, column에 추가하여 자료 손실을 방지
liverpool_df['Player Name'] = liverpool_df.index
liverpool_df 
```

```python
liverpool_df.set_index('Number', inplace = True)
liverpool_df
# 인덱스는 겹치는 값이 없는 것으로 하는게 좋다
```
