```python
import pandas as pd
```

```python
df = pd.read_csv('data/silicon_valley_details.csv')
df
```

```python
df['company'].value_counts()
```

```python
df['job_category'].value_counts()
```

```python
boolean_adobe = df['company'] == 'Adobe'
boolean_count = df['count'] > 0
boolean_race = df['race'] == 'Overall_totals'
boolean_job = (df['job_category'] != 'Totals') & (df['job_category'] != 'Previous_totals')

condition = boolean_adobe & boolean_count & boolean_race & boolean_job
```

```python
df_adobe = df[condition]
```

```python
df_adobe.set_index('job_category', inplace = True)
```

```python
df_adobe
```

```python
df_adobe.plot(kind = 'pie', y = 'count')
```
